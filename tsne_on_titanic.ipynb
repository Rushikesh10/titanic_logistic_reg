```python
import matplotlib.pyplot as plt 
import seaborn as sns
import sklearn
import pandas as pd
import numpy as np
titanic= sns.load_dataset("titanic")
```

```python
from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder()
ohe_town=enc.fit_transform(titanic["embark_town"].values.reshape(-1,1))
```

```python
print(ohe_town)
```

```python

titanic["alive"]=titanic["alive"].replace(to_replace='yes',value=1)
titanic["alive"]=titanic["alive"].replace(to_replace='no',value=0)
titanic["adult_male"]=titanic["adult_male"].replace(to_replace=True,value=1)
titanic["adult_male"]=titanic["adult_male"].replace(to_replace=False,value=0)
titanic["sex"]=titanic["sex"].replace(to_replace='male',value=1)
titanic["sex"]=titanic["sex"].replace(to_replace='female',value=0)
titanic["who"]=titanic["who"].replace(to_replace='man',value=1)
titanic["who"]=titanic["who"].replace(to_replace='woman',value=2)
titanic["who"]=titanic["who"].replace(to_replace='child',value=3)
titanic=titanic.drop(columns=["class","embarked","deck","alone"],axis=1)




```

```python
titanic.head()
```

```python
categories=list(titanic['embark_town'].values)
cat_list = []
for i in categories:
    i=str(i)
    temp = ""
    # consider we have text like this "Math & Science, Warmth, Care & Hunger"
    for j in i.split(','): # it will split it in three parts ["Math & Science", "Warmth", "Care & Hunger"]
        if 'The' in j.split(): # this will split each of the catogory based on space "Math & Science"=> "Math","&", "Science"
            j=j.replace('The','') # if we have the words "The" we are going to replace it with ''(i.e removing 'The')
        j = j.replace(' ','') # we are placeing all the ' '(space) with ''(empty) ex:"Math & Science"=>"Math&Science"
        temp+=j.strip()+" " #" abc ".strip() will return "abc", remove the trailing spaces
        temp = temp.replace('&','_') # we are replacing the & value into 
    cat_list.append(temp.strip())
from collections import Counter
my_counter = Counter()
for word in titanic['embark_town'].values:
    word=str(word)
    my_counter.update(word.split())
```

```python
titanic['clean_categories'] = cat_list
titanic.drop(['embark_town'], axis=1, inplace=True)
titanic.head(2)
```

```python
cat_dict = dict(my_counter)
sorted_cat_dict = dict(sorted(cat_dict.items(), key=lambda kv: kv[1]))
```

```python
for i, j in sorted_cat_dict.items():
    print("{:20} :{:10}".format(i,j))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(vocabulary=list(sorted_cat_dict.keys()), lowercase=False, binary=True)
vectorizer.fit(titanic['clean_categories'].values)
print(vectorizer.get_feature_names())


categories_one_hot = vectorizer.transform(titanic['clean_categories'].values)
print("Shape of matrix after one hot encodig ",categories_one_hot.shape)
```
